```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
attrition_df.nunique()
```

```python
y_df = attrition_df[["Attrition", "Department"]]
y_df
```

```python
# Create a list of at least 10 column names to use as X data
x = ["PercentSalaryHike",
"PerformanceRating",
"RelationshipSatisfaction",
"StockOptionLevel" ,
"TotalWorkingYears",
"TrainingTimesLastYear",
"WorkLifeBalance",
"YearsAtCompany",
"YearsInCurrentRole",
"YearsSinceLastPromotion",
"YearsWithCurrManager"
]



# Create X_df using your selected columns
X_df = attrition_df[x]

# Show the data types for X_df
X_df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_df, y_df)
X_train
```

```python
scaler = StandardScaler()

# Fit the StandardScaler to the training data
scaler.fit(X_train)

# Scale the training and testing data
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_train_scaled.shape
```

```python
OHE = OneHotEncoder(sparse=False)

# Fit the encoder to the training data
Y_train_department = Y_train["Department"].values.reshape(-1,1)
Y_test_department = Y_test["Department"].values.reshape(-1,1)
OHE.fit(Y_train_department)

# Create two new variables by applying the encoder
# to the training and testing data
Y_train_department_enc = OHE.transform(Y_train_department)
Y_test_department_enc = OHE.transform(Y_test_department)
Y_test_department_enc
```

```python
OHE_att = OneHotEncoder(sparse=False)
Y_train_attrition = Y_train['Attrition'].values.reshape(-1,1)
Y_test_attrition = Y_test['Attrition'].values.reshape(-1,1)
# Fit the encoder to the training data
OHE_att.fit(Y_train_attrition)

# Create two new variables by applying the encoder
# to the training and testing data
Y_train_att_encoded = OHE_att.transform(Y_train_attrition)
Y_test_attrition_encoded = OHE_att.transform(Y_test_attrition)
Y_test_attrition_encoded
```

```python
# Find the number of columns in the X training data
column_num = X_train.shape[1]

# Create the input layer
input_layer = layers.Input(shape=(column_num, ), name='input_layer')

# Create at least two shared layers
Dense_1 = layers.Dense(64, activation='relu')(input_layer)
Dense_2 = layers.Dense(128, activation='relu')(Dense_1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
Department_dense = layers.Dense(64, activation='relu')(Dense_2)

# Create the output layer

Department_output = layers.Dense(3, activation='softmax', name='Department_output')(Department_dense)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
Attrition_Dense = layers.Dense(64, activation='relu')(Dense_2)

# Create the output layer
Attrition_output = layers.Dense( 2, activation='sigmoid', name= 'Attrition_output')(Attrition_Dense)
```

```python
# Create the model
model = Model(inputs = input_layer, outputs=[
    Department_output,
    Attrition_output
])

# Compile the model
model.compile(optimizer='adam',
              loss= {'Department_output': 'categorical_crossentropy',
                     'Attrition_output': 'binary_crossentropy'},
                metrics = {'Department_output': 'accuracy',
                           'Attrition_output': 'accuracy'})

# Summarize the model
model.summary()
```

```python
# Train the model

model.fit(X_train_scaled, {
    'Department_output': Y_train_department_enc,
    "Attrition_output": Y_train_att_encoded
},
epochs =  100,
#batch_size = 10,
validation_split = 0.2)
```

```python
# Evaluate the model with the testing data
model_results = model.evaluate(np.array(X_test_scaled), {
    "Department_output": Y_test_department_enc,
    "Attrition_output": Y_test_attrition_encoded})
model_results
```

```python
print(f"Department predictions accuracy: {model_results[3]}")
print(f"Attrition output acccuracy: {model_results[4]}")
```
